```python
import vcf
import pandas as pd
from pylab import plot, show, savefig, xlim, figure, hold, ylim, legend, boxplot, setp, axes
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

###Start off by making a dictionary (called vcf_files)...
###...with the key as the vcf file and the value as the relevant column (population frequency)

```python

vcf_files = {'afr_47306.vcf':'AFR_FREQ', 'amr_47306.vcf':'AMR_FREQ', 'eas_47306.vcf':'EAS_FREQ','eur_47306.vcf':'EUR_FREQ', 'sas_47306.vcf':'SAS_FREQ'}
# for pop in [, 'EAS_FREQ','SAS_FREQ','ENSEMBL_FREQ',,'EUR_FREQ']:

#make new empty dictionary for results
results = {}
#start = True means we are only working on the first key-value pair
#we do this because we want to use the format from this to set the format, and then we'll just add data from other files
#otherwise we'd repeat the formatting / columns rather than adding to the existing format
start = True
for vcf_file in vcf_files:
    vcf_reader = vcf.Reader(open('1000genome/'+vcf_file, 'r'))
    if start:
        for record in vcf_reader:
            if record.ID is not None:
                result = {}
                result['chrom']= record.CHROM
                result['pos']=record.POS
                result['ref']=record.REF
                result['alt']=record.ALT[0]            
                if float(record.INFO[vcf_files[vcf_file]][0]) >0.5:
                    result[vcf_files[vcf_file]] = 1 - float(record.INFO[vcf_files[vcf_file]][0])
                else:
                    result[vcf_files[vcf_file]]=float(record.INFO[vcf_files[vcf_file]][0])
                    results[record.ID] = result
        start = False
    else:
        for record in vcf_reader:
            if record.ID is not None:
                if record.ID in results:
                    result = results[record.ID]
                    #we only want results[record.ID] (i.e rs values) and the rest of the columns are already there from before
    #                 result['chrom']= record.CHROM
    #                 result['pos']=record.POS
    #                 result['ref']=record.REF
    #                 result['alt']=record.ALT[0]            
                    if float(record.INFO[vcf_files[vcf_file]][0]) >0.5:
                       result[vcf_files[vcf_file]] = 1 - float(record.INFO[vcf_files[vcf_file]][0])
                #use float here to convert it to a number from a string
                    else:
                        result[vcf_files[vcf_file]]=float(record.INFO[vcf_files[vcf_file]][0])
                        results[record.ID] = result
                #this following bit is for any rs values that weren't in the first key:value pair of the dictionary (i.e they weren't in the AFR vfc)        
                else:
                    result = {}
                    result['chrom']= record.CHROM
                    result['pos']=record.POS
                    result['ref']=record.REF
                    result['alt']=record.ALT[0]            
                    if float(record.INFO[vcf_files[vcf_file]][0]) >0.5:
                        result[vcf_files[vcf_file]] = 1 - float(record.INFO[vcf_files[vcf_file]][0])
                    else:
                        result[vcf_files[vcf_file]]=float(record.INFO[vcf_files[vcf_file]][0])
                        results[record.ID] = result
```
